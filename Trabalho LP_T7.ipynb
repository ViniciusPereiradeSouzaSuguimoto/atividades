```r
library(ggplot2)
library(tidyverse)
```

```r
data = read.csv('https://github.com/Rogerio-mack/Visualizacao-de-Dados-em-Python/raw/main/data/Life_Expectancy_Data.csv')
head(data)
```

```r
str(data)
head(data)
```

```r
data %>% select(Life.expectancy,Adult.Mortality,Population,Country )  %>%
    head(100)
```

```r
library(RColorBrewer)
```

```r
ggplot(data, aes(x=Year, y=Life.expectancy)) + 
 geom_point(col="darkcyan", size=10) 
```

```r
ggplot(data, aes(x=Year, y=Life.expectancy))+ geom_col(size=1) +
  ggtitle("Analise de Expectativa de vida", subtitle="Ano X Expectativa de Vida") +
  xlab("Ano") +
  ylab("Vida.expectativa")
```
